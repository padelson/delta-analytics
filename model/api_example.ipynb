```python
from label_topics import *
import pandas as pd
import numpy as np
import requests
```

```python
# to get the hand-made categories need to download 
# word2vec from https://github.com/mmihaltz/word2vec-GoogleNews-vectors
import gensim
w2v_model = gensim.models.KeyedVectors.load_word2vec_format('../GoogleNews-vectors-negative300.bin', binary = True)
```

```python
# Example using OMF API
url = "https://open.ompnetwork.org/api/captions?start=0&limit=500"
data = requests.get(url).json()
caps = [x['caption'] for x in data['results']]
sess = [x['session_id'] for x in data['results']]
```

Keep in mind this API only returns correct results when the full text of a session is included in the output. That's why below we see many examples of sessions with empty labels because the current OMF API doesn't guarantee that results are returned in groups of sessions.

```python
# Example when you pass raw output form OMF API
lt = LabelTopics(data=data)
lt.clean_data()
lt.predict_fluff()
lt.predict_topic_labels()
lt.get_w2v_predictions(w2v_model, session_thresh=.7)
lt.combine_preds()
lt.topic_labels
```

```python
# Passing in lists of captions and sessions
test_caps = [x['caption'] for x in data['results']]
test_sess = [x['session_id'] for x in data['results']]

lt = LabelTopics(captions=test_caps, session_ids=test_sess)
lt.clean_data()
lt.predict_fluff()
lt.predict_topic_labels()
lt.get_w2v_predictions(w2v_model, session_thresh=.7)
lt.combine_preds()
lt.topic_labels
```
